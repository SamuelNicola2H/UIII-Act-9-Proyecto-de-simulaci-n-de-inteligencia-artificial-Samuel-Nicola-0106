```python
# UIII Act 9: Proyecto de simulación de inteligencia artificial
# Alumno: Samuel Nicola
# Rol asignado: Entrenador

print("==================================================")
print("  ¡Bienvenido a EntrenadorIA - Tu Coach Virtual!  ")
print("==================================================")

# Lista inicial de atletas vacía, tal como se solicitó
entrenador = []

# Bandera para mantener el ciclo del menú activo
ejecutando = True

while ejecutando:
    # Menú interactivo y atractivo para preparatoria
    print("\n--- MENÚ DE CONTROL DEL ENTRENADOR IA ---")
    print("1. Registrar nuevo atleta en el equipo")
    print("2. Ver lista de atletas entrenando")
    print("3. Consultar recomendación de la IA para el equipo")
    print("4. Vaciar lista de atletas (Reiniciar equipo)")
    print("5. Salir del simulador")

    opcion = input("\nSelecciona una opción (1-5): ")

    # Opción 1: Registrar un atleta (Simulación de entrada de datos a la IA)
    if opcion == "1":
        nuevo_atleta = input("Introduce el nombre del atleta a registrar: ")
        entrenador.append(nuevo_atleta)
        print("🤖 [IA]: Atleta '" + nuevo_atleta + "' registrado con éxito y asignado a su plan de entrenamiento.")

    # Opción 2: Mostrar los atletas actuales en la lista
    elif opcion == "2":
        if entrenador == []:
            print("🤖 [IA]: El equipo está vacío. ¡Registra a alguien en la opción 1!")
        else:
            print("🏃‍♂️ Atletas actualmente en el equipo:")
            print(entrenador)

    # Opción 3: Simulación de Inteligencia Artificial (Toma de decisiones con if/elif/else)
    elif opcion == "3":
        print("🤖 [IA] Analizando el rendimiento del equipo...")

        # Contamos cuántos atletas hay usando la comparación de la lista
        # (Sin usar ciclos for ni funciones complejas)
        cantidad = len(entrenador)

        if cantidad == 0:
            print("⚠️ Alerta IA: No hay atletas. No puedo generar recomendaciones sin datos.")
        elif cantidad <= 2:
            print("💡 Recomendación IA: El equipo es pequeño. Sugiero un entrenamiento personalizado de alta intensidad.")
        elif cantidad <= 5:
            print("💡 Recomendación IA: Tamaño de equipo ideal. Buen momento para tácticas grupales y resistencia.")
        else:
            print("⚠️ Alerta IA: ¡Demasiados atletas! El entrenador necesita apoyo. Divide el grupo en dos turnos.")

    # Opción 4: Resetear la lista (Uso de condicionales para limpiar)
    elif opcion == "4":
        if entrenador == []:
            print("🤖 [IA]: La lista ya se encuentra vacía.")
        else:
            entrenador = []
            print("💥 Sistema reiniciado. Todos los atletas han sido removidos de la lista.")

    # Opción 5: Salir del programa rompiendo el ciclo while
    elif opcion == "5":
        print("\n🤖 [IA]: Guardando datos de entrenamiento... ¡Buen trabajo hoy, Coach Samuel!")
        print("Fin de la simulación.")
        ejecutando = False

    # Validación por si el usuario teclea algo que no está en el menú
    else:
        print("❌ Opción no válida. Por favor, elige un número del 1 al 5.")
```
